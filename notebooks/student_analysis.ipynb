```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
%%time

path="../data/student_depression_dataset.csv"
df = pd.read_csv(path)
```

```python
%%time

target_column="Depression"
sns.countplot(x=target_column, data=df)
plt.title("Depression Distribution")
plt.show()
```
