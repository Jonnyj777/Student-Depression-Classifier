```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
%%time

path="../data/student_depression_dataset.csv"
df = pd.read_csv(path)
```

```python
%%time

target_column = "Academic Pressure"
group_column = "Depression"

# Compute the average
grouped = df.groupby(group_column)[target_column].mean().reset_index()

# Plot the average
sns.barplot(data=grouped, x=group_column, y=target_column)
plt.title("Average Academic Pressure by Depression Status")
plt.ylabel("Average Academic Pressure")
plt.xlabel("Depression")
plt.show()
```
